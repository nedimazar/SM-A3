```python
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
# Getting our predictions from earlier
in_domain = pd.read_csv('data/in_domain_hateBERT.csv')
cross_domain = pd.read_csv('data/cross_domain_hateBERT.csv')
```

```python
# Extracting the true labels and predictions
in_domain_y = in_domain['labels']
in_domain_pred = in_domain['predictionns']

# Extracting the true labels and predictions
cross_domain_y = cross_domain['labels']
cross_domain_pred = cross_domain['predictions']
```

```python
# Generating the classification report
print(classification_report(y_true=in_domain_y, y_pred=in_domain_pred))
```

```python
# Generating the classification report
print(classification_report(y_true=cross_domain_y, y_pred=cross_domain_pred))
```

```python
# Drawing a confusion matrix
ConfusionMatrixDisplay.from_predictions(in_domain_y, in_domain_pred, display_labels=['NOT', 'OFF'], cmap='PuBuGn')
```

```python
# Drawing a confusion matrix
ConfusionMatrixDisplay.from_predictions(cross_domain_y, cross_domain_pred, display_labels=['NOT', 'OFF'], cmap='PuBuGn')
```
